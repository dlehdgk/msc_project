```python
import numpy as np
import scipy.special as sc
import matplotlib.pyplot as plt
```

```python
#constants
T0 = 2.7255
z3 = sc.zeta(3)
pi = np.pi
eta = 6.05e-10
e4 = np.exp(4)

#Gaussian units
q = 4.8032e-10
m = 9.1095e-28
c=3e10
kb=1.3803e-16
hbar=1.0546e-27
hc = 6.6262e-27*c
B = m*q**4/(2*hbar**2)
a0 = hbar**2/(m*q*q)
sig0 = 512*(pi*q*a0)**2/(3*e4*hbar*c)
A = sig0*c/np.sqrt(pi)*(q*q/(hbar*c))**3
r0 = q*q/(m*c*c)
#1eV in kelvin
ev = 1.6022e-12/kb
```

```python
#functions

#Temperature
def Temp(z):
    return T0*(1+z)
def xn(n, T):
    return B/(n*n*kb*T)
#integral of e^x E1(x) from 0 to x
def inte(x):
    return np.exp(x)*sc.exp1(x)+0.577+np.log(x)
#lower bound (integral from n = 2 -> infty)
def lb(T):
    x2 = xn(2,T)
    return = A*np.sqrt(x2)*inte(x2)
#upper bound (integral from n =1 -> infty)
def ub(T):
    x1 = xn(1,T)
    return 0.5*A*np.sqrt(x1)*inte(x1)
#trapezium approximation
def trap(T):
    x2 = xn(2,T)
    a2 = 0.5*A*x2*np.sqrt(x2)*np.exp(x2)*sc.exp1(x2)
    return a2+lb(T)
#Padmanabhan volume 3 approximation
def Pad3(T):
    return 2.6e-13*(ev/T)**0.8
#Padmanabhan cosmology review
def rev(T):
    ratio = B/(kb*T)
    return 9.78*r0*r0*c*np.sqrt(ratio)*np.log(ratio)
    
    

def a_sum(T, n):
    tot = 0
    for i in range(2,n+1):
        xn = B/(i*i*kb*T)
        tot += xn**1.5*np.exp(xn)*sc.exp1(xn)
    return A*tot

def Pad(T):
    x2 = B/(4*kb*T)
    t1 = x2*np.exp(x2)*sc.exp1(x2)*(0.5+1/x2)
    return A*np.sqrt(x2)*(t1+0.577+np.log(x2))

def vol3(T):
    return 2.6e-13*(T*kb/1.6022e-12)**-0.8

def inte(T):
    x2 = B/(4*kb*T)
    t1 = np.exp(x2)*sc.exp1(x2)+np.log(x2)+0.577
    return A*np.sqrt(x2)*t1

def gs_inte(T):
    x1 = B/(kb*T)
    t1 = np.exp(x1)*sc.exp1(x1)+np.log(x1)+0.577
    return A*np.sqrt(x1)*t1
```

```python
print(Pad(1e4))
print(a_sum(1e4,25))
print(vol3(1e4))
print(inte(1e4))
print(gs_inte(1e4))
```

```python
T = np.linspace(1000, 2000, 500)
a10 = a_sum(T, 10)
a5 = a_sum(T, 5)
P = Pad(T)
P3 = vol3(T)
I = inte(T)
Igs = gs_inte(T)
```

```python
plt.plot(T, P, label='Volume 1')
plt.plot(T, P3, label='volume 3')
plt.plot(T, a10, label='n = 10')
plt.plot(T, a5, label='n = 5')
plt.plot(T, I, label='integral')
plt.plot(T, Igs, label='ground state')
plt.grid()
plt.legend()
plt.xlabel('Temperature [K]')
plt.ylabel('$\\alpha$ [$cm^3\cdot s^{-1}$]')
plt.show()
```

```python
def func(x):
    n1 = (1/(x+1))**2
    n = (1/x)**2
    top = n**3*np.exp(n1)*sc.exp1(n1)
    bottom = n1**3*np.exp(n)*sc.exp1(n)
    return top/bottom
x = np.linspace(1e4, 1e10,100)
y = func(x)
```

```python
plt.plot(x,y)
plt.grid()
plt.xscale('log')
#plt.yscale('log')
plt.show()
```
