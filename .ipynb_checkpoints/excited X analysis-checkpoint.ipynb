```python
from constants import *
import matplotlib.pyplot as plt
```

```python
def ba(T):
   return (m*kb*T/(2*pi*hbar**2))**1.5*np.exp(-B/(4*kb*T)) 
```

```python
T = np.linspace(1e3, 1e6)
y = ba(T)
plt.plot(T, y)
plt.xscale('log')
plt.ylabel('$\\beta/\\alpha$')
plt.xlabel('Temperature [K]')
plt.grid()
plt.show()
```
