```python
from constants import *
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#constants
#alternative photoionisation x-section
#sig0 = 64*q*q*pi*a0**2/(3*np.sqrt(3)*hbar*c)
print(B/1.6022e-12)
print(a0)
```

```python
#Temperature in erg!!!
def Temp(z):
    return T0*(1+z)
#n=1 only my derivation

#note temperatures below are in Kelvin!!
def a1(T):
    return A*np.sqrt(B/(kb*T))

#nth state only full
def an(T, n):
    xn = B/(n*n*kb*T)
    return A*xn**1.5*np.exp(xn)*sc.exp1(xn)

#from caltech
def cal(T):
    return 4.2e-13*(1.6022e-12/(kb*T))**0.7

#approximation of sum to infty
def recombA(T):
    x1 = B/(kb*T)
    t1 = (x1+1)*np.exp(x1)*sc.exp1(x1)+np.log(x1)+0.577
    return 0.5*A*np.sqrt(x1)*t1

#manual sum from state n to state m
def a_sum(T, n, m):
    tot = 0
    for i in range(n, m+1):
        tot += an(T, i)
    return tot

#approximation for large x for recomb A
def a_bigx(T):
    x1 = B/(kb*T)
    t1 = (x1+1)/x1+np.log(x1)+0.577
    return 0.5*A*np.sqrt(x1)*t1
```

```python
#find T = 1eV
test = ev/kb
print('Temperature at 1 eV is: {} K'.format(test))
```

```python
print(a1(test))
print(recombA(test))
print(a_sum(1e4, 1, 10))
```

```python
print("percentage difference between alpha_1 and Padmanabhan value is: {} %".format((a1(test)-1.4e-13)/1.4e-13*100))
```

```python
#z = np.linspace(100,10000, 500)
#graph of alpha for padmanabhan, my derivation, and caltech value
T = np.linspace(100, 10000, 1000)
y = a1(T)
y2 = 1.4e-13*np.sqrt(ev/(kb*T))
y3 = cal(T)

plt.plot(T, y, label='equation 2.18')
plt.plot(T, y2, label='equation 2.17')
plt.plot(T, y3, label='cal (2011)')
plt.ylabel('$\\alpha$ [$cm^3\cdot s^{-1}$]')
plt.xlabel('Temperature [K]')
plt.xscale('log')
plt.grid()
plt.legend()
plt.savefig('Images/alpha_gs.png', dpi=300)
plt.show()
```

```python
fact = recombA(test)
t = np.linspace(5.0e2, 1.0e4, 10000)
aA = recombA(t)
y_fact = aA/fact
y_5 = (test/t)**0.5
y_7 = (test/t)**0.7
```

```python
plt.plot(t, y_fact, label='$\\alpha_A$')
plt.plot(t, y_7, label='$\propto 10^{-0.7}$')
plt.plot(t, y_5, label='$\propto 10^{-0.5}$')
plt.xlabel('Temperature [K]')
plt.xscale('log')
plt.ylabel('$\\alpha/\\alpha(T = 1 eV)$')
plt.legend()
plt.grid()
plt.savefig('Images/comparison of T dept gs.png', dpi=300)
plt.show()
```

```python
#percentage difference at T = 10^3
f1 = recombA(1e3)/fact
f2 = (test/1e3)**0.7

print('percentage difference is:', (f2-f1)/f1*100)
```

```python
#fitting function for aA
#fit function
def fit_func(x, a, b):
    #a = coefficient at 1 eV
    #b = temperature at 1 eV
    #c = power to fit
    return a*(test/x)**b
p0 = np.array([4.2e-13, 0.7])
po,co = curve_fit(fit_func, t, aA, p0)
print(po, np.sqrt(co[0,0]), np.sqrt(co[1,1]))
```

```python
fit = fit_func(t, po[0], po[1])
```

```python
plt.plot(t, fit, label='fit')
plt.plot(t, aA, label='$\\alpha_A$')
plt.grid()
plt.legend()
plt.show()
```

```python
#over range of z = [800, 1300]
z = np.linspace(800,1300, 1000)
te = Temp(z)/kb
fit1 = fit_func(te, po[0], po[1])
fit2 = recombA(te)
```

```python
plt.plot(z, fit1, label='fit')
plt.plot(z, fit2, label='$\\alpha_A$')
plt.grid()
plt.legend()
plt.show()
```
