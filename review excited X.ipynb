```python
from constants import *
import matplotlib.pyplot as plt
import scipy.optimize as op
from scipy.integrate import odeint
from numba import jit
```

```python
#functions
#Temperature in Kelvin
@jit(nopython=True)
def Temp(z):
    return T0*(1+z)
@jit(nopython=True) 
def n_g(z):
    return (Temp(z)/hc)**3*16*pi*z3
@jit(nopython=True) 
def Hu(z):
    H0 = 100*h/3.0857e19
    return H0*np.sqrt(Orh2/h/h*(1+z)**4+Omh2/h/h*(1+z)**3+Ol)
@jit(nopython=True)
def alph(z):
    return 9.78*r0**2*c*np.sqrt(B/Temp(z))*np.log(B/Temp(z))
@jit(nopython=True)
def beta(z):
    return (m*Temp(z)/(2*pi*hbar**2))**1.5*np.exp(-B/Temp(z))
@jit(nopython=True)
def diff(z, X):
    amp = alph(z)/((1+z)*Hu(z))
    term1 = beta(z)*(1-X)
    term2 = eta*n_g(z)*X*X
    return -amp*(term1-term2)
@jit(nopython=True)
def solver(func, x, y0):
    y = np.zeros_like(x)
    y[0] = y0
    for i in range(len(x)-1):
        y[i+1] = y[i]+(x[i+1]-x[i])*func(x[i],y[i])
    return y
@jit(nopython=True)
def X_eq(z):
    S = 4*np.sqrt(2/pi)*z3*eta*(Temp(z)/(m*c*c))**1.5*np.exp(B/Temp(z))
    X = (-1+np.sqrt(1+4*S))/(2*S)
    return X
@jit(nopython=True)
def rev_ex(z):
    amp = 2.4e-3*np.sqrt(Omh2)/0.02207
    return amp*(z/1000)**12.75
```

```python
#plotting diff(z, X)
z = np.linspace(800, 1400, 50)
X = np.linspace(0,1, 50)
x,y = np.meshgrid(z, X)
z = diff(x, y)
```

```python
fig = plt.figure(constrained_layout=True)
ax = plt.axes(projection='3d')
ax.plot_surface(x,y,z)
ax.set_xlabel('redshift')
ax.set_ylabel('ionisation fraction')
ax.set_zlabel('dX/dt')
ax.grid()
plt.savefig('Images/dx_by_dt.png', dpi=300)
plt.show()
```

```python
#solve diff(z, X)
#start at z = 1700, X = 1
#dz< 0 Xi+1 = Xi -|dz|*dx/dz
#go in increments of dz = 5
red = np.arange(1700.0, 800.0, -1e-5)
ion = solver(diff, red, 1.0)
Saha = X_eq(red)
```

```python
plt.plot(red,ion, label='excited states solution')
plt.plot(red, Saha, label='Saha solution')
plt.grid()
plt.legend()
plt.xlabel('redshift')
plt.ylabel('ionisation fraction')
plt.show()
```

```python
t = np.linspace(1700.0, 800.0, 500)
sol = odeint(diff, X_eq(1700), t, tfirst=True)
#print(sol)
```

```python
s = sol[:,0]
saha = X_eq(t)
trunc = np.linspace(800, 1200, 100)
review_fit = rev_ex(trunc)
delta = sol[:,0]-X_eq(t)
```

```python
plt.plot(t, s, label='excited states solution')
plt.plot(t, saha, label='Saha solution')
plt.plot(trunc, review_fit, label='$\propto z^{12.75}$')
plt.grid()
plt.xlabel('redshift')
plt.ylabel('ionisation fraction')
plt.yscale('log')
plt.legend()
plt.savefig('Images/excited.png', dpi=300)
plt.show()
```
